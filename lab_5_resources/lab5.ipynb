```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
R = np.arange(1, 1000+1, 1)
Q = np.arange(-500, 500+2, 1)
R, Q = np.meshgrid(R, Q)
Vth = 3
Rth = 500 - 200j
RL = R+Q*1j
VL = (Vth*RL)/(Rth+RL)
I = VL/RL
P = (VL*np.conjugate(I))
Rthideal = np.real(((Vth*np.conjugate(Rth))/(np.conjugate(Rth)+Rth))*(((Vth*np.conjugate(Rth))/(np.conjugate(Rth)+Rth))/(np.conjugate(Rth))))

plt.figure()
ax = plt.axes(projection='3d')
power = ax.plot_surface(R, Q, np.real(P),cmap=cm.coolwarm)
ax.scatter(np.real(Rth),-np.imag(Rth),Rthideal, color='green')
ax.text(np.real(Rth),-np.imag(Rth), Rthideal, "$R_{ideal}$")
ax.view_init(45, -65, 0)
ax.set_ylabel("j")
ax.set_xlabel("R")
ax.set_zlabel("P (W)")
plt.colorbar(power, shrink=0.5, orientation="horizontal")
plt.savefig("power_vs_R_J.png")
```
