```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dt = pd.read_csv("example_csv.csv") # import CSV file
headers = list(dt.columns.values) # get the names of each column
dt = dt.iloc[1:-1] # clean off the first row
dt = dt.to_numpy().astype( dtype=np.float32).T # convert to a numpy array and transpose it, make sure data is a float32
print("Data shape: ")
print(dt.shape)
```

```python
t = dt[0] # specify the time column

# plot each sin graph

# voltage
plt.figure()
for i in range(1, dt.shape[0]-4):
    plt.plot(t, dt[i], label=f"{headers[i]}") # lable each line

plt.ylabel("Voltage (V)")
plt.xlabel("time (t)")
plt.legend(bbox_to_anchor=(1, 1), ncol=1)
plt.grid()
plt.savefig("voltages.png")


# current
plt.figure()
for i in range(dt.shape[0]-3, dt.shape[0]):
    plt.plot(t, dt[i], label=f"{headers[i]}") # lable each line

plt.ylabel("Current (I)")
plt.xlabel("time (t)")
plt.legend(bbox_to_anchor=(1, 1), ncol=1)
plt.savefig("currents.png")
plt.grid()
```
